```python
import pandas as pd
from  sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
Dataset=pd.read_csv("Churn.csv")
```

```python
Dataset.head()
```

```python
Dataset.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
Dataset['NumOfProducts'].unique()
```

```python
Dataset['Geography'].unique()
```

```python
Dataset['Gender'].unique()
```

```python
Encode_gender=LabelEncoder()
```

```python
Dataset['Gender']=Encode_gender.fit_transform(Dataset['Gender'])
```

```python
with open('LabelEncoder_Gender.pkl','wb') as file1:
    pickle.dump(Encode_gender,file1)
```

```python
Dataset
```

```python
ohe_geo=OneHotEncoder()
```

```python
Dataset
```

```python
Encoded_Data=ohe_geo.fit_transform(Dataset[['Geography']])
```

```python
Features=ohe_geo.get_feature_names_out()
```

```python
ohe_data=Encoded_Data.toarray()
```

```python
df=pd.DataFrame(ohe_data,columns=Features)
```

```python
Dataset=pd.concat([Dataset.drop('Geography',axis=1),df],axis=1)
```

```python
Dataset
```

```python
with open('OneHotEncoder_Geography.pkl','wb') as file2:
    pickle.dump(ohe_geo,file2)
```

```python
X=Dataset.drop('Exited',axis=1)
```

```python
y=Dataset['Exited']
```

```python
X
```

```python
scaler=StandardScaler()
```

```python
X_train,X_test,y_train,y_tesr=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
X_train_Scaled=scaler.fit_transform(X_train)
```

```python
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_Scaled
```

```python
with open('Scaler.pkl','wb') as file2:
    pickle.dump(scaler,file2)
```

```python
X_train_Scaled.shape[1]
```

ANN 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train_Scaled.shape[1],)),#First Hidden Layer Connected with  I/P Layer
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')# Output Layer

])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()

```

```python
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

Settingup TensorBoard

```python
import datetime
```

```python
log_dir='logs/fit/'+ datetime.datetime.now().strftime("%d/%m/%Y-%H:%M:%S")
```

```python
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    X_train_Scaled,y_train,validation_data=(X_test_scaled,y_tesr),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
